```python
import pandas as pd
import numpy as np

df = pd.read_csv("MER_T09_03.csv")

df.head()

df.Description.unique()

df.Description = df.Description.apply(lambda x: x.replace('Landed Cost of Crude Oil Imports From ',''))

df.Description.unique()

data = df

data = data.drop(columns=['MSN','Column_Order','Unit'])
data.head()
data.columns = ['year','Price', 'Country']
data.dtypes

data.year= data.year.apply(lambda x: str(x))

data = data[data['year'].str.endswith('13')]
data.head()

data.year= data.year.apply(lambda x: int(x[:-2]))

data.head()

data = data[data.year>=1986]

data.Price = data.Price.astype('float', errors='ignore')

data1 = data[(data.Country == 'All Non-OPEC Countries') | (data.Country =='All OPEC Countries')]
data1.head()

drop_indices = np.array(np.where((data.Country == 'All Non-OPEC Countries') | (data.Country =='All OPEC Countries'))[0])
data = data.reset_index()
data = data.drop(drop_indices, axis=0)
data = data.drop(columns=['index'])
data.Country.unique()
data = data.drop(np.where(data.Country == 'Persian Gulf Nations')[0])
data.to_csv('price_imported.csv', index=False)
data1.to_csv('No_OPEC_prices.csv')
```
