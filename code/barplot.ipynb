```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('MER_T03_03C.csv')
```

```python
df.head()
```

```python
df.Description = df.Description.apply(lambda x: x.replace('Petroleum Imports From ',''))

df.Description.unique()

df = df[df.YYYYMM>=197301]

data = pd.DataFrame(columns=['date']+ list(df.Description.unique()))

for col in data.columns[1:]:
    #indices = np.where(df.Column_Order==(i+1) & (df.YYYYMM>=197301))[0]
    data[col] = np.array(df.Value[df.Description==col])
    data[col] = pd.to_numeric(data[col], errors='coerce')

data['date'] = np.array(df.YYYYMM[df.Description==data.columns[1]])

data.head()

data.shape

data.dtypes

data['date']= data.date.apply(lambda x: str(x))

data = data[~data['date'].str.endswith('13')]

data.head()

data['date'] = data['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m', errors='coerce'))

data.dtypes

data['year'] = data.date.dt.year

data = data[['date','year']+list(df.Description.unique())]

data.head()

data.columns

data.columns = ['date', 'year', 'Algeria', 'Angola',
       'Ecuador', 'Iraq', 'Kuwait', 'Libya', 'Nigeria',
       'Saudi Arabia', 'Venezuela', 'Other OPEC', 'Total OPEC']

data.head()
```

```python
data.to_csv('imports_OPEC.csv', index=False)
```

```python
df_exports = pd.read_csv('MER_T03_03F.csv')

df_exports = df_exports[df_exports.YYYYMM>=198401]

df_exports.Description = df_exports.Description.apply(lambda x: x.replace('Petroleum Exports to ',''))

df_exports.YYYYMM = df_exports.YYYYMM.apply(lambda x: str(x))

df_exports = df_exports[~df_exports.YYYYMM.str.endswith('13')]

#df_exports.YYYYMM = df_exports.YYYYMM.apply(lambda x: x[:-2])

df_exports.YYYYMM = df_exports.YYYYMM.apply(lambda x: pd.to_datetime(str(x), format='%Y%m', errors='coerce'))
df_exports['month']= df_exports.YYYYMM.dt.month

df_exports.YYYYMM = df_exports.YYYYMM.apply(lambda x: str(x))
df_exports.YYYYMM = df_exports.YYYYMM.apply(lambda x: x[:4])
df_exports['Value'] = pd.to_numeric(df_exports['Value'], errors='coerce')
```

```python
df_exports.dtypes
```

```python
df_exports.Value = df_exports.Value/1000
```

```python
df_exports.to_csv('exports_monthly.csv')
```
